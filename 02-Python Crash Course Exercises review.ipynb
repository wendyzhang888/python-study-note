**How can I return both the sum and product of two numbers in Python?**
- You must complete the task by using function

```python
def sum_and_product(x,y):
    """This function takes two numbers and returns their sum and product."""
    a=x+y
    b=x*y
    return a,b

number1 = 5
number2 = 7

# Using a single variable to hold the returned tuple
c,d=sum_and_product(number1,number2)
print(c,d)
```

**You are driving a little too fast, and a police officer stops you. Write a function to return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket". If your speed is 60 or less, the result is "No Ticket". If speed is between 61 and 80 inclusive, the result is "Small Ticket". If speed is 81 or more, the result is "Big Ticket". Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 higher in all cases.**

```python
def caught_speeding(speed, is_birthday):
    if is_birthday:
        speed=speed-5
        if speed<=60:
            return "No Ticket"
        elif speed<=80:
            return "Small Ticket"
        else:
            return "Big Ticket"
    else:
        if speed<=60:
            return "No Ticket"
        elif speed<=80:
            return "Small Ticket"
        else:
            return "Big Ticket"
a=caught_speeding(81,False)
print(a)
```
