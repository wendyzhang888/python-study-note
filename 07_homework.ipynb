```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
plt.figure(figsize=(12,7))


plt.subplot(1,2,1)
plt.scatter(df.index,df['total_bill'])
plt.xlabel('Index')
plt.ylabel('Total bill')
plt.title('Index vs Total bill')


plt.subplot(1,2,2)
plt.hist(df['total_bill']) 
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.title(' histogram total bill')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['total_bill'], df['tip'], color='blue', alpha=0.6,s=3)

# Labels and title
plt.xlabel('Total bill')
plt.ylabel('tip')
plt.title('Scatter Plot of total bill vs tip')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['tip'], df['size'], color='blue', alpha=0.6,s=3)

# Labels and title
plt.xlabel('Tip')
plt.ylabel('Size')
plt.title('Scatter Plot of Tip vs Size')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['tip'], df['price_per_person'], color='blue', alpha=0.6,s=3)

# Labels and title
plt.xlabel('tip')
plt.ylabel('price_per_person')
plt.title('Scatter Plot of Index vs Total Bill')

# Show the plot
plt.show()
```
