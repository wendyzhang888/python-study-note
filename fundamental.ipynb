```python
def sum(x,y):
    result=x+y
    print(result)
sum(3,4)
sum(1,8)#按照位置传参
```

```python
def sum(x,y):
    result=x+y
    print(result)
sum(y=1,x=7)#指定传参
```

```python
def sum(x,y):
    result=x+y
    return 10
num=sum(3,1)
print(num)#return后面是啥，啥就被sum替代
```

```python
def square(x):
    result=x*x
    return result

num=square(16)
print(num)
```

```python
def area(x,y):
    result=x*y
    return result#return，def是固定词，其他所有都是variable，可以换任何名字

num=area(3,10)
print(num)
```
